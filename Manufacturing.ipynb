```python
# Pass_Fail	Time	start_time
x = result.drop(['Pass_Fail','Time','start_time','year'], axis=1)
y = result[['Pass_Fail']]
```

```python
x.head()
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
```

```python
x = pd.DataFrame(x_scaler, columns=x.columns[:])
```

```python
x.head()
```

```python
#pip install imblearn
```

```python
import imblearn
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
x_ros, y_ros = ros.fit_resample(x, y)
from collections import Counter
print(Counter(y))
print("##############"*5)
print(Counter(y_ros))
```

```python
y_ros.value_counts()
```

```python
x_ros.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(.95)
x_pca = pca.fit_transform(x_ros)
x_pca.shape
```

```python
x_pca
```

# Splitting Dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_pca, y_ros, test_size=0.3, random_state=1)
```

# Building Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score 
from sklearn.metrics import f1_score, precision_score, recall_score   
```

```python
log = LogisticRegression()
log.fit(x_train, y_train)
```

```python
y_pred_train_log = log.predict(x_train)
y_pred_test_log = log.predict(x_test)
```

```python
print(confusion_matrix(y_train, y_pred_train_log))
print("##########"*10)
print(confusion_matrix(y_test, y_pred_test_log))
```

```python
print(classification_report(y_train, y_pred_train_log))
print("##########"*10)
print(classification_report(y_test, y_pred_test_log))
```

```python
print("Training Accuracy :", accuracy_score(y_train, y_pred_train_log))
print("##########"*10)
print("Test Accuracy :", accuracy_score(y_test, y_pred_test_log))
```

```python
rforest = RandomForestClassifier(n_estimators=500, criterion='entropy')
rforest.fit(x_train, y_train)
```

```python
y_pred_train = rforest.predict(x_train)
y_pred_test = rforest.predict(x_test)
```

```python
print(confusion_matrix(y_train, y_pred_train))
print("##########"*10)
print(confusion_matrix(y_test, y_pred_test))
```

```python
print(classification_report(y_train, y_pred_train))
print("##########"*10)
print(classification_report(y_test, y_pred_test))
```

```python
print("Training Accuracy :", accuracy_score(y_train, y_pred_train))
print("##########"*10)
print("Test Accuracy :", accuracy_score(y_test, y_pred_test))
```
